- **Selenium:**

```python
import time
from selenium.webdriver import Firefox
from selenium.webdriver import ActionChains
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.remote.command import Command
from selenium.webdriver import Firefox
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By 
```

```python


#operating in headless mode
opts = Options()
opts.headless = True
browser = Firefox(options=opts)
```

```python
browser.get('https://www.bolsamadrid.es')
```

```python

# buscamos el link  "MERCADOS Y COTIZACIONES"
 
mercados_link = browser.find_element_by_link_text("MERCADOS Y COTIZACIONES")
time.sleep(2)
mercados_link.click()

acciones = WebDriverWait(browser, 5).until(EC.element_to_be_clickable((By.LINK_TEXT,"Acciones")))
acciones.click()
```

```python
# definimos la tabla

time.sleep(2)
table = browser.find_element(By.ID, "ctl00_Contenido_tblAcciones")

columns=[]

for element in table.find_elements(By.TAG_NAME, "th"): # encontramos las categorias   
    columns.append(element.text)
    
print(columns)
```

```python
results=[]

for element in table.find_elements(By.TAG_NAME, "td"):   # encontramos todos los datos.
    results.append(element.text)

print(results)
```

```python
data = {}
for i in range(8):
    data[columns[i]]=results[i::len(columns)]## creamos el diccionario

df = pd.DataFrame(data)
df
```
